```python
import requests
import bs4
import string
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
state = "Nordrhein-Westfalen"
#job_title = "Softwareentwickler (m/w/d)"
job_title_set = ("Softwareentwickler", "Softwareentwickler-Java", "Senior-Softwareentwickler", 
                 "IT-Systems-Engineer", "Java-Entwickler")

max_results_per = 10

columns = ["state", "full_job_title", "company_name", "url"]

df = pd.DataFrame(columns = columns)
df.head()

```

```python
for title in job_title_set:
    for state in state:
        #print(title)
        #print(state)
        for start in range(0, 200, max_results_per):
            #https://de.indeed.com/Jobs?q=Softwareentwickler-Java&l=Nordrhein-Westfalen
            url = f"https://de.indeed.com/Jobs?q={title}&l=Nordrhein-Westfalen&start={start}"
            #print(url)
            #print(start)
            page = requests.get(url)
            time.sleep(1+random.random()*4)  #ensuring at least 1 second between page grabs
            soup = BeautifulSoup(page.text, from_encoding="utf-8")
            job_post = [] 

            for div in soup.find_all(name="div", attrs={"class":"row"}):
                #specifying row num for index of job posting in dataframe
                num = (len(df) + 1) 
                #creating an empty list to hold the data for each posting

                #append state name
                job_post.append("Nordrhein-Westfalen")
                #df.loc['state'][num]="Nordrhein-Westfalen"
                #grabbing job title
                for a in div.find_all(name="a", attrs={"data-tn-element":"jobTitle"}):
                    #print(a["title"])
                    job_post.append(a["title"])
                    
                #grabbing company name
                company = div.find_all(name="span", attrs={"class":"company"}) 

                if len(company) > 0: 
                    for b in company:
                        job_post.append(b.text.strip()) 
                else: 
                    sec_try = div.find_all(name="span", attrs={"class":"result-link-source"})
                    for span in sec_try:
                        job_post.append(span.text) 
                        
                #grabbing location name
                #c = div.findAll('span', attrs={'class': 'location'})

                #for span in c:
                    #job_post.append(span.text)    
                    
                #grabbing job url
                for a in div.find_all(name="a", attrs={"data-tn-element":"jobTitle"}):
                        #print(a["href"])
                        job_post.append("https://de.indeed.com" + a["href"]) 
            #print(job_post)
                #df.loc[num] = job_post
                #print(df)
            #job_post.append(title)   
            #print(job_post)    
                
        
```

```python
job_company_description = []
for url in job_company_url:
    #print(df.iloc[i,0:3])
    #url = df.iloc[i,3]
    #print(url)
    page = requests.get(url)
    time.sleep(1+random.random()*10)  #ensuring at least 1 second between page grabs
    soup = BeautifulSoup(page.text, from_encoding="utf-8")
    
    for div in soup.find_all(name="div", attrs={"class":"jobsearch-jobDescriptionText"}):
        job_company_description.append(div.text)
        #df.iloc[i,5] = div.text
```
